In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:

- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
# Your code goes here
import pandas as pd
df1 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
df2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
df3 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")
```

```python
print(f"{df1.shape}, {df2.shape}, {df3.shape}")
```

```python
# Clean dataset 1
clean_columns = [col.replace(' ', '_').lower() for col in list(df1.columns)]
clean_columns[1] = "state"
df1.columns = clean_columns
print(df1.shape)
df1.head()
```

```python
# Clean dataset 2
clean_columns = [col.replace(' ', '_').lower() for col in list(df2.columns)]
clean_columns[1] = "state"
df2.columns = clean_columns
print(df2.shape)
df2.head()
```

```python
# Clean dataset 3
clean_columns = [col.replace(' ', '_').lower() for col in list(df3.columns)]
clean_columns[1] = "state"
df3.columns = clean_columns
print(df3.shape)
df3.head()
```

```python
df = pd.concat([df1, df2, df3], axis=0).reset_index(drop=True)
print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
#  Clean column "customer_lifetime_value"
df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace("%", "")
df['education'].value_counts()
```

```python
# Clean column "education"
df['education'] = df['education'].str.replace("Bachelors", "Bachelor")
df['education'].value_counts()
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Your code goes here
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv")
df.head()
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
df['sales_channel'].value_counts()
```

```python
# Create pivot table: total revenue by sales channel
pivot_sales = pd.pivot_table(
    df,
    index='sales_channel',                   
    values='customer_lifetime_value',       
    aggfunc='sum'
)
# Round the total revenue to 2 decimal points
pivot_sales = pivot_sales.round(2)
```

```python
# Sort descending to see top channel
pivot_sales = pivot_sales.sort_values(by='customer_lifetime_value', ascending=False)
pivot_sales
```

Some insights:
- Agent generated the most revenue.
- Web or online sales contribute the least revenue.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
pivot_gender_edu = pd.pivot_table(
    df,
    index='gender',                        
    columns='education',                   
    values='customer_lifetime_value',     
    aggfunc='mean'                        
)
pivot_gender_edu
```

Some insights:
- For Bachelor and High School or Below, females have slightly higher average customer lifetime value than males. 
- For College, Doctor, and Master, males have slightly higher customer lifetime value than females.
- However, the differences between them are very small.
